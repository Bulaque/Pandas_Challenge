# PyCity Schools Analysis

- Charter schools have a lower school budget but have sizebly higher passing rates (roughly 90%) while District schools have a higher school budget and a very noticable drop in passing rates (roughly 53% passing). This shopws that there is a strong negative correlation with school budget and Overall passing grades.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_csv = Path("Resources/schools_complete.csv")
student_data_csv = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

# Combine the data into a single dataset.  
merged_school_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
merged_school_data.head(2920)
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = merged_school_data["school_name"].nunique()
print(school_count)
```

```python
# Calculate the total number of students
student_count_a = merged_school_data["Student ID"].max()
student_count = student_count_a + 1
print(student_count)
```

```python
# Calculate the total budget
total_budget = merged_school_data["budget"].sum()
print(total_budget)
```

```python
# Calculate the average (mean) math score
added_math_score = merged_school_data["math_score"].sum()
average_math_score = (added_math_score) / 39170
print(round(average_math_score,2))
```

```python
# Calculate the average (mean) reading score
added_reading_score = merged_school_data["reading_score"].sum()
average_reading_score = (added_reading_score) / 39170
print(round(average_reading_score,2))
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = merged_school_data[(merged_school_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(round( passing_math_percentage,2))
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = merged_school_data[(merged_school_data["reading_score"] >=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
print(round(passing_reading_percentage,2))
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = merged_school_data[
    (merged_school_data["math_score"] >= 70) & (merged_school_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(round(overall_passing_rate,2))
```

```python
District_Summary_DF = pd.DataFrame({"Total number of schools": [school_count],
                                    "Total Students": [student_count],
                                    "Total Budget": [total_budget],
                                    "Average Math Score": [average_math_score],
                                    "Average Reading Score": [average_reading_score],
                                    "Percent Passing Math": [passing_math_percentage],
                                    "Percent Passing Reading": [passing_reading_percentage],
                                    "Percent Overall Passing": [overall_passing_rate]})
District_Summary_DF
```

## School Summary

```python
# Use the code provided to select all of the school types
# Set Indexing, refer to type of school separatley (end of day 2 or 3)

school_types = school_data.set_index("school_name")["type"]
school_types.head(10)
```

```python
# Calculate the total student count per school
    # There is a function we learned about when we learned about creating DFs

students_per_school = merged_school_data["school_name"].value_counts()
print(students_per_school)
```

```python
# Calculate the total school budget and per capita spending per school

per_school_budget = merged_school_data.groupby("school_name")["budget"].mean()
per_school_budget
per_school_capita = per_school_budget/students_per_school
per_school_capita
```

```python
# Calculate the average test scores per school - Math

avg_math = merged_school_data.groupby("school_name")["math_score"].mean()
avg_math
```

```python
# Calculate the average test scores per school - Reading
avg_reading = merged_school_data.groupby("school_name")["reading_score"].mean()
avg_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher

math_70 = merged_school_data[merged_school_data["math_score"] >= 70]
school_students_passing_math = math_70.groupby("school_name").size()
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher

reading_70 = merged_school_data[merged_school_data["reading_score"] >= 70]
school_students_passing_reading = reading_70.groupby("school_name").size()
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = merged_school_data[
    (merged_school_data["reading_score"] >= 70) & (merged_school_data["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
print(school_students_passing_math_and_reading)
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / students_per_school * 100
per_school_passing_reading = school_students_passing_reading / students_per_school * 100
overall_passing_rate = school_students_passing_math_and_reading / students_per_school * 100
print(overall_passing_rate)
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
# Include the following: School name, School type, Total students, Total school budget, Per student budget, Average math score, Average reading score, % passing math (the percentage of students who passed math), % passing reading (the percentage of students who passed reading), % overall passing (the percentage of students who passed math AND reading, per_school_summary = 
per_school_summary = pd.DataFrame({
    "School Type": school_types, 
    "Total Students": students_per_school,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame

per_school_summary



# Plug in each variable separately to see which variables work
# The index should be the same for each variable
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=labels)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame ({
    "Math Scores": round(size_math_scores,2),
    "Reading Scores": round(size_reading_scores,2),
    "Passing Math Scores": round(size_passing_math,2),
    "Passing Reading Scores": round(size_passing_reading,2),
    "Overall Passing Scores": round(size_overall_passing,2),
})

# Display results
size_summary
```
